We build the DAG from Fig 3.11

```python
G = CausalModel()
G.add_causation('X', 'Y')
G.add_causation('X', 'Z')
G.add_causation('Z', 'Y')

G.pos = {'X': (0, 0), 'Z': (1, 1), 'Y': (2, 0)}
```

```python
fig, ax = plt.subplots(1, figsize=(2.2,2.2))
G.plot(ax=ax)
```

```python
G.save_model('dags/Primer.Fig.3.11.dot')
```

And the DAG from Fig 3.12

```python
G = CausalModel()
G.add_causation('X', 'Y')
G.add_causation('X', 'Z')
G.add_causation('Z', 'Y')
G.add_causation('I', 'Y')
G.add_causation('I', 'Z')

G.pos = {'X': (0, 0), 'Z': (1, 1), 'Y': (2, 0), 'I': (2, 1)}
```

```python
fig, ax = plt.subplots(1, figsize=(2.2,2.2))
G.plot(ax=ax)
```

```python
G.save_model('dags/Primer.Fig.3.12.dot')
```

We can easily compute the modified graph resulting from intervening on X and Z:

```python
G2 = G.intervention_graph(['X', 'Z'])
fig, ax = plt.subplots(1, figsize=(2.2,2.2))
G2.plot(ax=ax)
```
